```python
import pandas as pd
import numpy as np
```

```python
df2016 = pd.read_csv('SP500_data_2016_close.csv',index_col=0)
df2017 = pd.read_csv('SP500_data_2017_close.csv',index_col=0)
df2018 = pd.read_csv('SP500_data_2018_close.csv',index_col=0)
df2019 = pd.read_csv('SP500_data_2019_close.csv',index_col=0)
df2020 = pd.read_csv('SP500_data_2020_close.csv',index_col=0)
df2021 = pd.read_csv('SP500_data_2021_close.csv',index_col=0)
```

```python
sectores = []
retornos = []
for sector in allcomp['Topic'].unique():
    sectores.append(sector)
    sec_returns = []
    df = pd.DataFrame(returns_df[allcomp[allcomp['Topic'] == sector].ticker.values.tolist()])
    for i in range(len(df)):
        diadf = pd.DataFrame(df.iloc[i])
        diadf.columns = ['Return']
        diadf['sim_sect'] = allcomp[allcomp['Topic'] == sector].sim_sect.values.tolist()
        total_weight = diadf.dropna()['sim_sect'].sum()
        diadf['TotalWeight'] = total_weight
        diadf['Rel_weight'] = diadf['sim_sect'] / diadf['TotalWeight']
        diadf['Rel_return'] = diadf['Return']* diadf['Rel_weight']
        dailyret = diadf['Rel_return'].sum().round(4)
        sec_returns.append(dailyret)
    retornos.append(sec_returns)
```

```python
retornos_etf = pd.DataFrame.from_records(retornos).transpose()
```

```python
sectores = [str(sec) for sec in sectores]
```

```python
retornos_etf.columns = sectores
```

```python
retornos_etf = retornos_etf + 1
```

```python
cotizaciones_etfs_new = retornos_etf.cumprod()
```
